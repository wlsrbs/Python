## Class 만들기

```python

class 클래스이름:
```

[규칙]==> 이 부분을 이해해야 함
* 파이썬 method는 항상 첫 번째 파라미터로 self 사용
  * 인자가 필요없을 때에도 self는 사용
* 클래스의 attribute는 내부에서 접근시, **self.attribute명 으로 접근**

````python
class Car:
    brend = ''
    name = ''
    color = ''
    
    def get_info(self):
        return f'회사:{self.brend},이름:{self.name},색상:{self.color}'
````

````python
mycar1 = Car()
mycar1
````

````python
mycar2 = Car()
mycar2
````

### 메서드의 매개변수

* 메서드의 첫 번째 매개변수는 항상 self!!!

````python
class Car:
    def get_info(self):
        return f'회사:{self.brend},이름:{self.name},색상:{self.color}'
    
    # 메서드의 첫 번째 매개변수는 항상 self!!!
    def set_data(self,brend,name,color):
        self.brend = brend
        self.name = name
        self.color = color
````

````python
mycar1 = Car()
mycar1.set_data('Hyundai','Grandure','White')
mycar1.get_info()
````

> * 매개변수를 정의할때는 4개, 호출할때는 3개? 
> * 메서드를 호출하면 첫 번째 매개변수 self는 객체 mycar1에 의해 자동으로 전달되기 때문

### 생성자(Constructor)

```python

def ___init___(self):

```

````python
mycar1 = Car()
mycar1.get_info()
````

> * set_data() 메서드를 수행하지 않고 get_info()메서드를 먼저 호출하면
> * 클래스내에 Attribute가 선언되지 않아 오류가 발생한다.
> * mycar1이라는 객체가 생성될 시점부터 초기값이 설정되어야 할 때 생성자를 사용한다. 
> * 즉, 객체가 생성될 때 자동으로 호출되는 메서드를 의미한다. 
> * [참고] __init__  언더바, 언더바로 이름을 가진 메서드는 내부 메서드의 경우 그렇게 선언함  



````python
class Car:
    def __init__(self,brend,name,color):
        self.brend = brend
        self.name = name
        self.color = color
    
    def get_info(self):
        return f'회사:{self.brend},이름:{self.name},색상:{self.color}'
    
    # 메서드의 첫 번째 매개변수는 항상 self!!!
    def set_data(self,brend,name,color):
        self.brend = brend
        self.name = name
        self.color = color
````

````python
mycar1 = Car()
````

> * 클래스를 객체로 만들때 사전에 정의 된대로 3개의 매개변수가 전달되지 않았다!! 라는 오류

````python
mycar1 = Car('Hyundai', 'Sonata', 'black')
mycar1.get_info()
````

## 연습문제

* 미션: 계산기를 만들어 봅시다. 
* 2개의 숫자에 대해 덧셈, 뺄셈, 곱셈, 나눗셈의 결과를 출력할 수 있음 


````python
class FourCal:
    def __init__(self,first,second):
        self.first = first
        self.second = second
    def set_data(self,first,second):
        self.first = first
        self.second = second
    def add(self):
        result = self.first + self.second
        return result
    def sub(self):
        result = self.first - self.second
        return result
    def mul(self):
        result = self.first * self.second
        return result
    def div(self):
        result = self.first / self.second
        return result
````

````python
calc = FourCal(4,3)

print(calc.add())
print(calc.sub())
print(calc.mul())
print(calc.div())
````
