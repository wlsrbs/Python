````python
s = 'silcing'

# 객체 = [start:end:] 
print(s[0:3:])  # 0 ~ 2까지 문자열 출력, 출력값: sil

# 객체 = [start::]
print(s[0::])  # 0 ~  s의 마지막 문자열까지 출력, 출력값: silcing

# 객체 = [:end:]
print(s[:4:])  # s 문자열 길이 3까지 출력, 출력값: silc
````

#### 문자열 포맷팅

* 문자열 안에 어떤 값을 삽입하는 방법
* 'OOO 고객님, 안녕하세요.'에서 OOO에 이름만 바꿔서 출력하려고 한다면...
* %d,%s 를 사용할 수 있고,

##### % 연산자로 포맷팅하기

````python
print("파이썬 시험 점수: %d 점" % 100) #숫자로해도
print("파이썬 시험 점수: %s 점" % 100) #문자열로 해도 되네!
````

````python
print("%s 님의 파이썬 시험 점수: %s 점" % ('박보검',100))
````

##### format()함수로 포맷팅하기

````python
print("파이썬 시험 점수: {} 점".format(100))
````

````python
print("{} 님의 파이썬 시험 점수: {:.1f} 점".format('박보검',92.53))
````

````python
name='박보검'
score=92.35
print(f"{name} 님의 파이썬 시험 점수: {score:.1f} 점")
````

### 리스트 자료형

* 숫자와 문자열 자료형은 변수하나에 한 가지 정보만 할당할 수 있습니다. 
* 정보를 좀 더 효율적으로 핸들링하기 위해 복합 자료형이 필요합니다. 
* 복합 자료형에는 리스트, 튜플, 딕셔너리, 집합이 있습니다. 
* 해당 자료형에 사용할 수 있는 함수의 종류가 너무 많아 익숙해 지는데 어려움이 있습니다. 
* 하지만
* 모든 자료구조(Data Structure)를 익히는데 관통하는 방법이 있습니다. 
* CRUD-LTS: Create(생성), Read(읽기), Update(수정), Delete(삭제)-len(길이),type(자료타입),sort(정렬)

#### 리스트란?

> 연관있는 데이터를 효율적으로 다룰 수는 없을까?

* 순서를 가진 복합 자료형으로 정수형, 실수형, 문자형 등 모든 자료들을 다양한 형태로 저장하고 활용할 수 있는 복합 자료구조

* \[   \] 안에 , 를 사용해서 데이터를 써넣으면 됨
    ```
    * 예: mydata = [1, 2, 'hello', 1.2]
    ```
* 각 **요소**는 **인덱스**를 사용해서 다룰 수 있음

### 튜플 자료형

* 튜플(Tuple)은 리스트와 거의 유사합니다. 
* 여러 값을 한꺼번에 모을 수 있습니다. 
* 튜플은 괄호()를 생략가능합니다. 
* 값을 변경/삭제할 수 없습니다. -> 프로그램중 고정된 값들을 모아두기 좋은 데이터 구조
* CRUD-LTS
    - 생성: 변수명 = (값1,값2,값3,...)
    - 읽기: 변수명[인덱스]
    - 수정: 요소 수정 불가
    - 삭제: 요소 삭제 불가

    - LTS:len(튜플),type(튜플),sorted(튜플)

### 딕셔너리 자료형

* 딕셔너리(Dictionary)는 관련된 정보를 서로 대응시켜 놓은 것입니다. 
* key(키)와 value(값)의 쌍으로 이루어져 있습니다. 
* 변수명 = {키1:값1, 키2:값2}
* 키로 값에 접근합니다. 
* CRUD-LTS
    - 생성: 변수명 = {키1:값1, 키2:값2}
    - 읽기: 변수명[키], 변수명.keys(), 변수명=values()
    - 수정: 변수명[키]=값
    - 삭제: del 변수명[키]

    - LTS:len(딕셔너리),type(딕셔너리),sorted(딕셔너리)

#### Create: 딕셔너리 생성

````python
data_dic = {'한국':'kr','일본':'jp','중국':'cn'}
data_dic
````

#### Read: 딕셔너리 읽기

* 키가 있을 때

````python
data_dic["한국"]
````

* 키가 없을 때

````python
data_dic['미국']
````

* 에러처리 귀찮으니 get(키) 함수 사용

````python
print(data_dic.get('한국'))
print(data_dic.get('미국'))
````

* 리스트처럼 슬라이스는 안 됨!

````python
data_dic['한국':'중국']
````

````python
data_dic['한국'] = 'KR'
data_dic
````

````python
data_dic['미국'] = 'US'

data_dic
````

#### Delete:딕셔너리 삭제
