### continue,break문

* continue: 반복을 할 때 조건을 걸어 실행하지 않고 건너뛰게 함
* break: 일정한 조건이 되면 모든 실행을 중지시킴

* 60점 이상인 사람에게는 축하 메시지 보내고, 나머지에게는 아무 메시지도 보내지 않는 프로그램 만들기

````python
marks = [90, 25, 67, 45, 80, 30, 90, 100]
number = 0

for mark in marks:
    number = number + 1
    if mark < 60:
        continue
    print(f"{number}번 학생 축하합니다. 합격입니다.")
````

* 위의 프로그램에서 5명(인덱스 5) 까지만 축하 메시지를 보내세요.

````python
marks = [90, 25, 67, 45, 80, 30, 90, 100]
number = 0

for mark in marks:
    number = number + 1
    if mark < 60:
        continue
    if number >= 6:
        break
    print(f"{number}번 학생 축하합니다. 합격입니다.")
````

### 리스트 내포

* 리스트 내포 (List Comprehension)을 사용하면 좀 더 편리하고 직관적인 프로그램을 만들 수 있습니다. 
* 리스트 안에 for문을 포함해서 만듭니다. 
* 리스트 내포는 파이썬스러운 코드 표현입니다. 

````python
a = [1,2,3,4]
result_2 = []
# 리스트에서 2곱하기
for num in a:
    result_2.append(num * 2) 
print(result_2)
````

````python
result_2 = []
result_3 = []
result_4 = []
result_5 = []

for num in a:
    result_2.append(num * 2) 
    result_3.append(num * 3) 
    result_4.append(num * 4) 
    result_5.append(num * 5) 

print(result_2)
print(result_3)
print(result_4)
print(result_5)

````

* 리스트 내포로 표현하기

````python
a = [1,2,3,4]
result_2 = [num * 2 for num in a]
result_3 = [num * 3 for num in a]
result_4 = [num * 4 for num in a]
result_5 = [num * 5 for num in a]
print(result_2)
print(result_3)
print(result_4)
print(result_5)
````

## While문

* for문이 횟수를 기준으로 반복하는 경우에 적합하다면,
* while문은 조건이 참일 때 반복하는 경우에 적합합니다. 
* while 조건식 뒤에 ':' 콜론을 꼭꼭꼭 사용합니다. 
* while 조건식이 참일 때 실행되는 실행문들은 꼭꼭꼭 들여쓰기를 해야 합니다.  

### 기본구조

```python
while 조건식:
    실행문1
    실행문2
    실행문3
```

````python
num = 0
while num < 10:
    num = num + 1
    print(num)
    
````

###  break문

* while문은 조건이 참인 동안 계속해서 반복실행합니다. 
* 이때, 강제로 빠져나가고 싶을 때 break문을 사용합니다. 

````python
참을성=10
월수입 = 1000

while 1000:
    
    참을성 -=1
    print('오늘도 일어납니다. 참을성:{}'.format(참을성))
    if 참을성 ==0:
        print("더이상 못 참겠습니다. 때려쳐!!")
        break
    
````
