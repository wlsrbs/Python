### 컬럼[본문 ] 내용 -> 하나의 문자열로 전환

### 한글만 추출

```python
import re
text_all = re.sub('[^ㄱ-ㅣ가-힣 ]',"",text_all) # 한글 아니 것(숫자, 특수문자, 영어) 제거
text_all
```

### 명사만 추출

```python
# !pip install konlpy
```

```python
from konlpy.tag import Okt
okt = Okt()
nouns_txt = okt.nouns(text_all)
nouns_txt
```

### 명사 빈도수 측정

```python
from collections import Counter     # jdk에서 제공하는 라이브러리
count = Counter(nouns_txt)          # Counter() : 카운트 하는 함수
count
```

```python
rank_text = count.most_common() # most_common()내림차순으로 정렬하는 함수
```

```python
rank_text
```

```python
len(rank_text)
```

### 상위 빈도수만 추출

```python
up_rank = []
for (keyword,num) in rank_text:
    
    if num > 8:
        up_rank.append((keyword,num))
    else:
        print(keyword,num)
```

```python
len( up_rank)
```

```python
up_rank
```

### 의미없는 단어 제거 ( 한글자 제거)
* 한글자여도 의미있는 단어가 있을 수 있으므로
* 수작업으로 진행

```python
up_rank
```

```python
# 한 글자로 된 데이터 삭제제
for keyword,num in up_rank:
    if len(keyword) == 1:
        print('up_rank.remove(("{}", {}))'.format(keyword,num))
```

```python
up_rank.remove(("제", 24))
up_rank.remove(("수", 21))
up_rank.remove(("때", 16))
up_rank.remove(("것", 16))
up_rank.remove(("책", 12))
up_rank.remove(("이", 12))
up_rank.remove(("요", 11))
up_rank.remove(("중", 10))
up_rank.remove(("저", 10))
up_rank.remove(("거", 9))
up_rank.remove(("좀", 9))
up_rank.remove(("왜", 9))
up_rank.remove(("뭐", 9))
```

```python
up_rank
```

```python
dic_rank = dict(up_rank)
dic_rank
```

### 워드클라우드 만들기

```python
# %pip install wordcloud
```

```python
#!pip install WordCloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
img_path = './image/bird_mask.png'
cloud_mask = np.array(Image.open(img_path))
```

```python
cloud_mask
```

```python
font = "/wsl.localhost/Ubuntu-22.04/usr/share/fonts/truetype/unfonts-core/UnDotumBold.ttf"
```

```python
# https://amueller.github.io/word_cloud/index.html
wordcloud = WordCloud(font_path=font, 
                      background_color="white", mask=cloud_mask)
wc = wordcloud.generate_from_frequencies(dic_rank)
plt.figure(figsize=(10,10))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()

## 워드크라우드 저장
wc.to_file('./image/wordcloud.png')
```
