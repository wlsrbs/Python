```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm  # 파일 경로 지시

print('버전 : ', mpl.__version__)
print('설치 위치: ', mpl.__file__)
print('설정 위치: ', mpl.get_configdir())
print('캐시 위치: ', mpl.get_cachedir())
print('설정파일 위치: ', mpl.matplotlib_fname()) # --> 글꼴 추가를 위해 경로로 이동하기 위함

font_list = fm.findSystemFonts(fontpaths=None, fontext='ttf')
# ttf 폰트 전체 개수
print(len(font_list))
```

```python
print(mpl.rcParams['font.family']) # 현재 폰트의 종류
print(mpl.rcParams['font.size'])   # 현재 폰트의 크기
```

```python
mpl.rcParams['font.family'] = 'NanumGothic' # 현재 폰트 기본값 변경
```

```python
#원하는 제목
target_col = ['After LIKE', 'Armageddon', '한 페이지가 될 수 있게', '해야 (HEYA)']

plt.rc('font',family='NanumGothic')   # 나눔고딕체를 사용할 경우
plt.rcParams['axes.unicode_minus']=False

plt.figure(figsize=(20,10))
#plt.plot(target_col)

plt.plot(music_pivot[target_col])

plt.legend(target_col, loc='best')
plt.gca().invert_yaxis() # y축 값 뒤집기
plt.xticks(rotation=90) # x 축 텍스트 90도 돌림
plt.xlabel("날짜", fontsize=12)     # x축 정보 표시
plt.ylabel("순위", fontsize=12)     # y축 정보 표시
plt.grid()
plt.show()
```

## 종합 코드

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

```python
music_df = pd.read_csv(f'./data/벅스탑100크롤링_20241016_20250124.csv')
music_df['순위'] = music_df['순위'].astype('int64')
music_pivot = pd.pivot_table(music_df,index='날짜',columns='제목',values='순위')

print('버전 : ', mpl.__version__)
print('설치 위치: ', mpl.__file__)
print('설정 위치: ', mpl.get_configdir())
print('캐시 위치: ', mpl.get_cachedir())  # 글꼴 추가 후, 해당 경로에 'file_list'파일 삭제 후 코드 재실행
print('설정파일 위치: ', mpl.matplotlib_fname()) # --> 글꼴 추가를 위해 경로로 이동하기 위함

mpl.rcParams['font.family'] = 'NanumGothic' # 현재 폰트 기본값 변경

#원하는 제목
target_col = ['After LIKE', 'Armageddon', '한 페이지가 될 수 있게', '해야 (HEYA)']

plt.rc('font',family='NanumGothic')   # 나눔고딕체를 사용할 경우
plt.rcParams['axes.unicode_minus']=False

plt.figure(figsize=(20,10))
#plt.plot(target_col)

plt.plot(music_pivot[target_col])

plt.legend(target_col, loc='best')
plt.gca().invert_yaxis() # y축 값 뒤집기
plt.xticks(rotation=90) # x 축 텍스트 90도 돌림
plt.xlabel("날짜", fontsize=12)     # x축 정보 표시
plt.ylabel("순위", fontsize=12)     # y축 정보 표시
plt.grid()
plt.show()
```
